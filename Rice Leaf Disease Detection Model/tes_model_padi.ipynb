```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt

# Load model (hanya perlu dilakukan sekali)
model = load_model('/content/model_padi.h5')

# Daftar label kelas
labels = [
    'rice_bacterial_leaf_blight',
    'rice_brown_spot',
    'rice_healthy',
    'rice_hispa',
    'rice_leaf_blast',
    'rice_leaf_scald',
    'rice_narrow_brown_spot',
    'rice_neck_blast',
    'rice_sheath_blight',
    'rice_tungro'
]

# Fungsi untuk prediksi dari path gambar
def prediksi_penyakit_daun(img_path):
    # Load dan preprocessing gambar
    img = load_img(img_path, target_size=(224, 224))
    img_array = img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Tampilkan gambar
    plt.imshow(img)
    plt.axis('off')
    plt.title("Gambar Uji")
    plt.show()

    # Prediksi
    predictions = model.predict(img_array)[0]
    predicted_index = np.argmax(predictions)
    confidence = predictions[predicted_index]
    predicted_label = labels[predicted_index]

    print(f"Prediksi: {predicted_label}")
    print(f"Confidence Score: {confidence:.2f}")
```

```python
prediksi_penyakit_daun('/content/padi-bercak.jpg')
```
